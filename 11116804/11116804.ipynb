```python
x = wine.density
y = wine.pH
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('density/g/cm^3')
plt.ylabel('pH')
plt.show()
```

### A Scatter Plot - Density vs pH

The scatter plot shows that there is a slight negative correlation between density and pH in the data. As the density increases, the pH levels tend to decrease. The correlation is not very strong but the downward trend is visible. This relationship makes intuitive sense - wines with higher density/sugar content tend to be more acidic with lower pH.
There are no major outliers in the data. The scatter plot shows clustering of points with densities between 0.99 and 1.00 g/cm^3 and pH between 3.0 and 3.4. This range contains a majority of the wines.
In summary, the scatter plot provides insight into the relationship between density and pH in the wine data. There are opportunities for further analysis by coloring the points based on other attributes, overlaying regression lines, or investigating specific clusters. But overall, the negative correlation aligns with domain knowledge of wine chemistry.

##  Histogram Diagrams

__Histograms were plotted for each of the features in the wine dataset to understand the distribution of the values. The histograms provide insight into the shape, central tendency and spread of the features.A histogram with 50 bins was plotted for each of the features in the wine dataset to understand the distribution of the variables.__

- The histogram for alcohol content shows a near normal distribution, with a peak around 10% alcohol. Most wines fall in the alcohol range of 8% to 14%. There are a few outliers on the lower end with alcohol < 8%.

- The histogram of density shows a strong right-skew distribution, with most wines having a density between 0.99 and 1.00 g/cm^3. Only a very small number of wines have a density outside this range.

- The sulfur dioxide histogram indicates two peaks - one around 20-40 and the other around 140-160. This is likely because sulfur dioxide is added at two stages in the wine making process. The bimodal distribution highlights that most wines have SO2 additions at one of those two levels.

- The citric acid histogram is right-skewed with a peak near 0 g/L. Most wines have low citric acid content of 0-1 g/L while only a few outliers have higher levels.

In summary, the histograms provide a statistical snapshot of the distribution of key variables. They highlight central tendencies, shapes, clusters and outliers for each feature. This analysis can inform deeper investigations into the impact of composition on wine quality.__

## Box and Whisker Plots 

__Boxplots were created for each of the 12 features in the wine dataset to visually inspect their distributions. The boxplots reveal several insights:__

- The median quality score is around 5 to 6, with a few high-end outliers. This suggests most wines are average, with some exceptional ones.

- Features like alcohol, pH, chlorides exhibit approximate normal distributions, clustered around the median.
- Citric acid, volatile acidity and sulphates are right-skewed, with most wines having low values.

- There are potential outliers, like wines with extreme alcohol, volatile acidity and chlorides values. Further investigation is required to assess if these are errors.

- No differentiation is observed between red and white wines. Both exhibit similar spreads, indicating composition is more indicative of quality than wine type.

In summary, the boxplots reveal the underlying distributions of the features. When combined with histograms, they provide a robust statistical snapshot and help identify relationships to explore further. Key non-normal features like citric acid and quality scores are apparent.
